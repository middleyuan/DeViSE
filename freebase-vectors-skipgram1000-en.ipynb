```python
import gensim
import numpy as np
```

```python
# load geogle pre-train word2vec embedding
model = gensim.models.KeyedVectors.load_word2vec_format('freebase-vectors-skipgram1000-en.bin', binary = True)
```

```python
CIFAR100_LABELS_LIST = [
    '/en/apple', '/en/fish', '/en/baby', '/en/bear', '/en/beaver', '/en/bed', '/en/bee', '/en/beetle', 
    '/en/bicycle', '/en/bottle', '/en/bowl', '/en/boy', '/en/bridge', '/en/bus', '/en/butterfly', '/en/camel', 
    '/en/can', '/en/castle', '/en/caterpillar', '/en/cattle', '/en/chair', '/en/chimpanzee', '/en/clock', 
    '/en/cloud', '/en/cockroach', '/en/couch', '/en/crab', '/en/crocodile', '/en/cup', '/en/dinosaur', 
    '/en/dolphin', '/en/elephant', '/en/flatfish', '/en/forest', '/en/fox', '/en/girl', '/en/hamster', 
    '/en/house', '/en/kangaroo', '/en/fingerboard', '/en/lamp', '/en/lawn_mower', '/en/leopard', '/en/lion',
    '/en/lizard', '/en/lobster', '/en/man', '/en/maple', '/en/motorcycle', '/en/mountain', '/en/mouse',
    '/en/mushroom', '/en/oak_tree', '/en/orange', '/en/orchid', '/en/otter', '/en/palm', '/en/pear',
    '/en/pickup_truck', '/en/pine', '/en/plain', '/en/plate', '/en/poppy', '/en/porcupine',
    '/en/possum', '/en/rabbit', '/en/raccoon', '/en/ray', '/en/road', '/en/rocket', '/en/rose',
    '/en/sea', '/en/seal', '/en/shark', '/en/shrew', '/en/skunk', '/en/skyscraper', '/en/snail', '/en/snake',
    '/en/spider', '/en/squirrel', '/en/streetcar', '/en/sunflower', '/en/pepper', '/en/table',
    '/en/tank', '/en/telephone', '/en/television', '/en/tiger', '/en/tractor', '/en/train', '/en/trout',
    '/en/tulip', '/en/turtle', '/en/wardrobe', '/en/whale', '/en/willow_tree', '/en/wolf', '/en/woman',
    '/en/worm'
]
maxtrix = np.zeros((100, 1000))
maxtrix = np.reshape(maxtrix,(100,1000))
```

```python
model['/en/maple']
```

```python
# get label embedding
string = model[str(CIFAR100_LABELS_LIST[0])]
string = string.reshape(1000,1)
```

```python
list(model.vocab.keys())[:10] 
```

```python
for i in range(100):
   string = model[str(CIFAR100_LABELS_LIST[i])]
   maxtrix[i] = string
maxtrix = np.reshape(maxtrix,(100*1000))
np.save('skipgram1000_label_embedding.npy',maxtrix)
```
